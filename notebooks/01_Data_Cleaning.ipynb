```python
import pandas as pd
import numpy as np
```

```python
#Load All Dataset Tables
base_url = "https://raw.githubusercontent.com/rizmiya-gith/NorthStar-Analytics/main/northstar_dataset/"

customers = pd.read_csv(base_url + "customers.csv")
orders = pd.read_csv(base_url + "orders.csv")
deliveries = pd.read_csv(base_url + "deliveries.csv")
complaints = pd.read_csv(base_url + "complaints.csv")
app_events = pd.read_csv(base_url + "app_events.csv")
drivers = pd.read_csv(base_url + "drivers.csv")
vehicles = pd.read_csv(base_url + "vehicles.csv")
incidents = pd.read_csv(base_url + "incidents.csv")
hubs = pd.read_csv(base_url + "hubs.csv")

print("All datasets loaded successfully!")
```

```python
#Verify Data Loaded
customers.head()
orders.head()
```

```python
#Check Dataset Shapes
print("Customers:", customers.shape)
print("Orders:", orders.shape)
print("Deliveries:", deliveries.shape)
print("Complaints:", complaints.shape)
print("App Events:", app_events.shape)
print("Drivers:", drivers.shape)
print("Vehicles:", vehicles.shape)
print("Incidents:", incidents.shape)
print("Hubs:", hubs.shape)
```

```python
#Missing Values Analysis (customer table)
print(customers.isnull().sum())
```

```python
#for all tables
tables = {
    "customers": customers,
    "orders": orders,
    "deliveries": deliveries,
    "complaints": complaints,
    "app_events": app_events,
    "drivers": drivers,
    "vehicles": vehicles,
    "incidents": incidents,
    "hubs": hubs
}

for name, table in tables.items():
    print(f"\n{name.upper()}")
    print(table.isnull().sum())
```

```python
#STEP 1 - DATA CLEANING
#Handle Missing Values
# Customers
customers["loyalty_score"].fillna(customers["loyalty_score"].mean(), inplace=True)
customers["preferred_channel"].fillna("Unknown", inplace=True)

# Complaints
complaints["compensation_amount"].fillna(0, inplace=True)

# Deliveries
deliveries["customer_rating_post_delivery"].fillna(
    deliveries["customer_rating_post_delivery"].mean(),
    inplace=True
)

# Drivers
drivers["training_score"].fillna(
    drivers["training_score"].mean(),
    inplace=True
)

# Vehicles
vehicles["battery_health_pct"].fillna(
    vehicles["battery_health_pct"].mean(),
    inplace=True
)

# Orders
orders["booking_channel"].fillna("Unknown", inplace=True)

print("Missing values handled successfully!")
```

```python
# DATA AUDIT

for name, table in tables.items():

    print("\n" + "="*50)
    print(f"TABLE: {name.upper()}")
    print("="*50)

    print("\nSHAPE:")
    print(table.shape)

    print("\nMISSING VALUES:")
    print(table.isnull().sum())

    print("\nDUPLICATES:")
    print(table.duplicated().sum())

    print("\nDATA TYPES:")
    print(table.dtypes)

    print("\nSUMMARY:")
    print(table.describe(include='all'))
```

```python
#1. Date Columns Are Still Objects
#2. Missing Values Still Exist
#3. Some Datatype Improvements Needed
#4. Inconsistent Zone Capitalization
#5. No Duplicate Problems
#6. No Impossible Numeric Values
```

```python
# DATE CONVERSION

customers["signup_date"] = pd.to_datetime(customers["signup_date"])

orders["order_created_at"] = pd.to_datetime(
    orders["order_created_at"]
)

deliveries["dispatch_time"] = pd.to_datetime(
    deliveries["dispatch_time"]
)

deliveries["delivery_completed_at"] = pd.to_datetime(
    deliveries["delivery_completed_at"]
)

complaints["created_at"] = pd.to_datetime(
    complaints["created_at"]
)

app_events["event_timestamp"] = pd.to_datetime(
    app_events["event_timestamp"]
)

incidents["reported_at"] = pd.to_datetime(
    incidents["reported_at"]
)

vehicles["commission_date"] = pd.to_datetime(
    vehicles["commission_date"]
)
```

```python
#STANDARDIZE ZONES
customers["home_zone"] = customers["home_zone"].str.upper()

orders["pickup_zone"] = orders["pickup_zone"].str.upper()

orders["dropoff_zone"] = orders["dropoff_zone"].str.upper()

drivers["base_zone"] = drivers["base_zone"].str.upper()

vehicles["assigned_zone"] = vehicles["assigned_zone"].str.upper()

hubs["zone"] = hubs["zone"].str.upper()
```

```python
#BOOLEAN CONVERSION
orders["special_handling_flag"] = (
    orders["special_handling_flag"].astype(bool)
)

app_events["success_flag"] = (
    app_events["success_flag"].astype(bool)
)

drivers["active_flag"] = (
    drivers["active_flag"].astype(bool)
)

deliveries["proof_of_completion_missing"] = (
    deliveries["proof_of_completion_missing"].astype(bool)
)
```

```python
# SAVE CLEANED DATASETS

customers.to_csv("customers_cleaned.csv", index=False)

orders.to_csv("orders_cleaned.csv", index=False)

deliveries.to_csv("deliveries_cleaned.csv", index=False)

complaints.to_csv("complaints_cleaned.csv", index=False)

app_events.to_csv("app_events_cleaned.csv", index=False)

drivers.to_csv("drivers_cleaned.csv", index=False)

vehicles.to_csv("vehicles_cleaned.csv", index=False)

incidents.to_csv("incidents_cleaned.csv", index=False)

hubs.to_csv("hubs_cleaned.csv", index=False)

print("All cleaned datasets saved successfully!")
```
